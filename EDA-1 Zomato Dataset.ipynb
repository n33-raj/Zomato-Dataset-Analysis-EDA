```python
sns.heatmap(df.isnull(),yticklabels = False,cmap='viridis')
plt.show()
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df,df_country,how="left",on='Country Code')
final_df.head()
```

```python
final_df.columns
```

```python
country_names= final_df.Country.value_counts().index
country_names
# value_counts() counts that how many times the same name is occuring in the rows.
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
#  top 3 countries that use Zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Observations: Zomato maximum records or transcations are from India after that USA and United Kingdoms

```python
# ratings= final_df[["Aggregate rating","Rating color","Rating text"]].value_counts().reset_index().rename(columns={0:"Rating Count"})
ratings = final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

### Observations:
1. when rating is between 4.5 - 4.9 -----> Excellent
2. when rating is between 4.0 - 4.4 -----> Very Good
3. when rating is between 3.5 - 3.9 -----> Good
4. when rating is between 2.5 - 3.4 -----> Average
5. when rating is between 1.8 - 2.4 -----> Poor


# 

```python
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(y="Rating Count",x="Aggregate rating",data=ratings)

plt.show()
```

```python
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(y="Rating Count",x="Aggregate rating",data=ratings,hue="Rating color",palette=["blue","red","orange","yellow","green","green"])

plt.show()
```

### Observations:
1. Not Rated count is very high
2. Maximum num of ratings are between 2.5 to 4.4
